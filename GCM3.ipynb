```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data= pd.read_csv('Responses)_new.csv')
```

```python
data.head()
```

```python
data=data.dropna(axis=1)
data.head()
```

```python
#data['Gender']=data['Gender'].map({'Male':0, 'Female':1})
data.drop(['Gender','Age'],axis=1,inplace=True)
```

```python
# Convert DataFrame to numeric type
data_numeric = data.apply(pd.to_numeric, errors='coerce')

# Drop rows where max value equals min value
data.drop(data_numeric.index[data_numeric.max(axis=1) == data_numeric.min(axis=1)].values, inplace=True)
```

```python
data.set_index(data.columns[0])
df=data.T
```

```python
df=df.drop(df.index[0])
df.head()
```

```python
# Construct the cause-effect matrix
cause_effect_matrix = df.values

# Calculate the influence matrix
influence_matrix = np.dot(cause_effect_matrix, cause_effect_matrix.T)
df1= pd.DataFrame(influence_matrix)
avg_matrix = (df1 - df1.min())/ (df1.max() - df1.min())
```

```python
avg_matrix
```

```python
avg_matrix = avg_matrix * 6 + 1
avg_matrix = avg_matrix.round().astype(int)  # Round to integers
avg_matrix
```

```python
causal_factors = np.where(avg_matrix.sum(axis=1) > avg_matrix.sum(axis=0))[0]

effect_factors = np.where(avg_matrix.sum(axis=1) < avg_matrix.sum(axis=0))[0]

print("Causal Factors:")
for factor in causal_factors:
    print(f"Variable {factor}")

print("\nConsequence Factors:")
for factor in effect_factors:
    print(f"Variable {factor}")
```

# Mathematical working step by step

```python
avg_matrix['Ri']= np.sum(avg_matrix,axis=1)
```

```python
avg_matrix.head()
```

```python
cj=np.vstack([avg_matrix,np.sum(avg_matrix,axis=0)])
avg_matrix_new= pd.DataFrame(cj)
```

```python
avg_matrix_new
```

```python
avg_matrix_new.rename(index={19: 'Cj'}, inplace=True)
avg_matrix_new.rename(columns={19: 'Ri'}, inplace=True)
```

```python
avg_matrix_new['Ri-Cj'] = avg_matrix_new.iloc[:, -1] - avg_matrix_new.iloc[-1, :]
```

```python
avg_matrix_new
```

```python
def func(value):
    if value<0:
        return 'Consequence'
    else: return 'Causal'

avg_matrix_new['Factors']= avg_matrix_new['Ri-Cj'].apply(func)
```
